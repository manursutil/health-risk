```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
bounds = {
    "trestbps": (80, 200),
    "chol": (100, 500),
    "thalach": (60, 210),
    "oldpeak": (0, 6.5)
}
```

```python
for col, (low, high) in bounds.items():
    df_heart_disease.loc[(df_heart_disease[col] < low) | (df_heart_disease[col] > high), col] = np.nan
```

```python
df_heart_disease.fillna(df_heart_disease.median(numeric_only=True), inplace=True)
```

```python
df_heart_disease.describe()
```

```python
heart_features = ["age", "trestbps", "chol", "thalach", "oldpeak"]
```

```python
X_heart = df_heart_disease[heart_features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_heart)
```

```python
pca = PCA()
```

```python
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         pca.explained_variance_ratio_.cumsum(),
         marker='o')
plt.title('Heart Disease PCA – Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.grid(True)
plt.savefig("/Users/manuelrodriguezsutil/Developer/health-risk/visuals/heart_disease/pca_heart.png", dpi=300)
plt.show()
```

```python
loadings = pd.DataFrame(
    pca.components_,
    columns=heart_features,
    index=[f'PC{i+1}' for i in range(len(heart_features))]
)
```

```python
print("Top PC1 Contributors:\n", loadings.loc['PC1'].sort_values(ascending=False))
print("\nTop PC2 Contributors:\n", loadings.loc['PC2'].sort_values(ascending=False))
```
